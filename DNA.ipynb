```python
# Function to convert ADN to ARNm
def ADN_to_ARNm(sequence):
    ARNm_sequence = ''
    for nucleotide in sequence:
        ARNm_sequence += ADN_to_ARNm_dict[nucleotide]
    return ARNm_sequence

ARNm_sequence = ADN_to_ARNm(ADN_sequence)
```

```python
# Define a dictionary for codons and their corresponding amino acids
codon_table = {
    "UUU": "Phe", "UUC": "Phe", "UUA": "Leu", "UUG": "Leu",
    "CUU": "Leu", "CUC": "Leu", "CUA": "Leu", "CUG": "Leu",
    "AUU": "Ile", "AUC": "Ile", "AUA": "Ile", "AUG": "Met",
    "GUU": "Val", "GUC": "Val", "GUA": "Val", "GUG": "Val",
    "UCU": "Ser", "UCC": "Ser", "UCA": "Ser", "UCG": "Ser",
    "CCU": "Pro", "CCC": "Pro", "CCA": "Pro", "CCG": "Pro",
    "ACU": "Thr", "ACC": "Thr", "ACA": "Thr", "ACG": "Thr",
    "GCU": "Ala", "GCC": "Ala", "GCA": "Ala", "GCG": "Ala",
    "UAU": "Tyr", "UAC": "Tyr", "UAA": "STOP", "UAG": "STOP",
    "CAU": "His", "CAC": "His", "CAA": "Gln", "CAG": "Gln",
    "AAU": "Asn", "AAC": "Asn", "AAA": "Lys", "AAG": "Lys",
    "GAU": "Asp", "GAC": "Asp", "GAA": "Glu", "GAG": "Glu",
    "UGU": "Cys", "UGC": "Cys", "UGA": "STOP", "UGG": "Trp",
    "CGU": "Arg", "CGC": "Arg", "CGA": "Arg", "CGG": "Arg",
    "AGU": "Ser", "AGC": "Ser", "AGA": "Arg", "AGG": "Arg",
    "GGU": "Gly", "GGC": "Gly", "GGA": "Gly", "GGG": "Gly",
}
```

```python
# Function to translate a genetic sequence into a protein
def translate_genetic_sequence(sequence):
    protein_sequence = ""
    for i in range(0, len(sequence), 3):
        codon = sequence[i:i+3]
        amino_acid = codon_table.get(codon, "-")
        if amino_acid == "STOP":
            break
        else:
            protein_sequence += amino_acid

    return protein_sequence
```

```python
# Translate the genetic sequence into a protein
protein_sequence = translate_genetic_sequence(ARN_sequence)

# Display the translated protein sequence
print("Translated Protein Sequence:", protein_sequence)
print("Length of Protein Sequence:", len(protein_sequence)/3)
```

```python
# Function to transcribe a genetic sequence into an RNA sequence
def transcribe_genetic_sequence(sequence):
    rna_sequence = sequence.replace("T", "U")
    return rna_sequence
```

```python
# Function to transcibe a genetic sequence into its complementary sequence
def complement_genetic_sequence(sequence):
    complement_sequence = ""
    for base in sequence:
        if base == "A":
            complement_sequence += "T"
        elif base == "T":
            complement_sequence += "A"
        elif base == "C":
            complement_sequence += "G"
        elif base == "G":
            complement_sequence += "C"
        else:
            complement_sequence += base
    return complement_sequence
```
